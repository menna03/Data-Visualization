```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import plotly.express as px
```

```python
ds = pd.read_csv("ds_salaries.csv")
```

```python
len(ds)
```

```python
ds.head()
```

First we want to see the comparison between the year and the salaries that are payed that year.
---



```python
sns.barplot(x="work_year", y="salary_in_usd", data=ds[:607])
```

We can conclude that the salaries grow over the years and it was at its most in 2022.

```python
sns.barplot(x="employment_type", y="salary_in_usd", data=ds[:607])
```

We can see also see the year compared to the experience level and we see that the experts has the most salary.



```python
sns.barplot(x="experience_level", y="salary_in_usd", data=ds[:607])
```

```python
g = sns.FacetGrid(ds, col="employment_type", height=3.5, aspect=.65)
g.map_dataframe(sns.barplot, x="experience_level",  y="salary_in_usd")
```

```python
sns.barplot(x="company_size", y="salary_in_usd", data=ds[:607])
```


We can see also that median and large companies has almost the same salary.

So we can see that the salary is related to the size of the copanies and experience level, and work year



```python
plt.figure(figsize=(20,10))
sns.barplot(x="company_location", y="salary_in_usd", data=ds[:607])
```

We can concluded that the best company  in salary is RU and we better not to go to IR or VN as it has the lowest salaries.

```python
plt.figure(figsize=(150,20))
#we need to do this size so that the names de clear
sns.barplot(x="job_title", y="salary_in_usd", data=ds[:607])
```

 We can see that the best salary in job title is “Dara analytics lead”.

```python
px.sunburst(ds, path=['work_year', 'company_size'], values='salary_in_usd', title=' salary compared to year and cmpany size')
```

We can see salary compared to year and company size. So, we can concluded that in 2022 most salaries came from medal companies and it was the most salary earned that year.
